## Finding Quarter Sales (By Default-USA)

```python
df['Quarter_USA']=df['Order Date'].dt.quarter
df
```

### Creating a Function For Indian Quarter

```python
def q(Quarter):
    if(Quarter==2):
        return 1
    elif(Quarter==3):
        return 2
    elif(Quarter==4):
        return 3
    else:
        return 4
```

```python
df['Quarter']=df['Quarter_USA'].apply(q)
df
```

## Now Removing the USA Quarter

```python
df=df.drop(columns='Quarter_USA')
df
```

## Sales According to Quarter

```python
ndf=df.groupby('Quarter').agg(
    Revenue=('Amount','sum'),
    Avg_Sale=('Amount','sum'),
    Quantity=('Quantity Ordered','sum')

)
ndf=ndf.reset_index()
ndf
```

## Line Chart

```python
plt.figure(figsize=(10,5))
ndf['Quarter']=ndf['Quarter'].astype(str)
plt.plot(ndf['Quarter'],ndf['Revenue'],marker='o',label='Retail Sales',color='green')
plt.title('Retail Sales')
plt.grid()
plt.legend()
plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.show()
```

## Bar Chart


```python
plt.figure(figsize=(10,5))
plt.bar(ndf['Quarter'],ndf['Revenue'],label='Retail Sales',color='blue',width=0.3)
plt.title('Retail Sales')
plt.grid()
plt.legend()
plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.xticks(rotation=30,fontsize=10)
plt.show()
```

## Bar Line Chart

```python
plt.figure(figsize=(10,5))
plt.plot(ndf['Quarter'],ndf['Revenue'],marker='s',label='Retail Sales',color='red')
plt.bar(ndf['Quarter'],ndf['Revenue'],label='Retail Sales',color='yellow',width=0.4)
plt.title('Retail Sales')
plt.grid()
plt.legend()
plt.xlabel('Quarter')
plt.ylabel('Sales')
plt.xticks(rotation=20)
plt.show()
```
